```python
from statsmodels.stats.power import TTestIndPower
import numpy as np
```

# **Initialize the Power Analysis Object**

```python
analysis = TTestIndPower()
```

# **Parameters**

```python
baseline_rate = 0.20
expected_rate = 0.18
effect_size = (baseline_rate - expected_rate) / np.sqrt(baseline_rate * (1 - baseline_rate)) # Cohen's h approximation
alpha = 0.05 # Type I error (5% chance of a false positive)
power = 0.80 # 1 - Type II error (80% chance of detecting the effect)
```

# **Calculate Sample Size**

```python
sample_size = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, ratio=1.0)
```

```python
print(f'ðŸš€ Product Recommendation: You need {int(np.ceil(sample_size))} users PER GROUP to detect a 2% lift in retention')
```
